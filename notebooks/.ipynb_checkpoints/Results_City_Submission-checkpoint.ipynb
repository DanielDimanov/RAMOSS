```python
import segmentation_models as sm
from tensorflow.keras.models import load_model

model = load_model('../Results/BEST-CITY.h5', compile=False)
```

```python
model.summary()
```

```python
miou = sm.metrics.IOUScore(threshold=0.5, per_image=True, class_indexes=[i for i in range(19)])
```

```python
from tqdm import tqdm
```

```python
sum_miou = 0
n_ims = 0
for b_im,b_m in tqdm(validation):
    output = model(np.array(b_im))
    im_miou = miou(b_m,output)
    sum_miou += im_miou
    n_ims += 1
```
